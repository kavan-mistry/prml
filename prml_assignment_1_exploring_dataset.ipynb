```python
import pandas as pd
```

* data loading

```python
d = pd.read_csv('E:\LABS\data_set\household_power_consumption.txt', sep=";")
df = pd.DataFrame(d)
df
```

* Read first 5 rows to get headers

```python
df.head(5)
```

* converting date and time variables to Date/Time format

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time

df
```

* Read 2900 rows that contains information on 2007-02-01 and 2007-02-02

```python
df_date = df.loc[(df['Date'] >= '2007-02-01') & (df['Date'] < '2007-02-03')].head(2900)
df_date
```

#### Exercise 2

* Subset the loaded data for 2007-02-01 and 2007-02-02

```python
subset_df = df.loc[(df['Date'] >= '2007-02-01') & (df['Date'] < '2007-02-03')]
subset_df
```

#### Exercise  3 : Create plots

* Histogram of Global active power consumption
* Global active power consumption over time

* Histogram of Global active power consumption

```python
df['Global_active_power'] = pd.to_numeric(df['Global_active_power'], errors='coerce')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df['Date'],df['Global_active_power'])
plt.xlabel('Date')
plt.ylabel('Global_active_power')
plt.title('Histogram of Global active power consumption')
# plt.rcParams['figure.figsize'] = [20, 10]
plt.show()
```

```python
df['Time'] = df['Time'].astype(str)
df
```

* Global active power consumption over time

```python
plt.plot(df['Time'], df['Global_active_power'])
plt.xlabel('Time')
plt.ylabel('Global_active_power')
plt.title('Global active power consumption over time')
plt.rcParams['figure.figsize'] = [10, 5]
plt.show()
```

* Energy sub metering

```python
df['Sub_metering_1'] = pd.to_numeric(df['Sub_metering_1'], errors='coerce')
df['Sub_metering_2'] = pd.to_numeric(df['Sub_metering_2'], errors='coerce')
df['Sub_metering_3'] = pd.to_numeric(df['Sub_metering_3'], errors='coerce')

plt.subplot(3, 1, 1)
plt.plot(df['Date'], df['Sub_metering_1'], label='Sub metering 1')
plt.xlabel('Date')
plt.ylabel('Sub-metering Values 1')

plt.subplot(3, 1, 2)
plt.plot(df['Date'], df['Sub_metering_2'], label='Sub metering 2')
plt.xlabel('Date')
plt.ylabel('Sub-metering Values 2')

plt.subplot(3, 1, 3)
plt.plot(df['Date'], df['Sub_metering_3'], label='Sub metering 3')
plt.xlabel('Date')
plt.ylabel('Sub-metering Values 3')

plt.rcParams['figure.figsize'] = [20, 20]
plt.show()
```
